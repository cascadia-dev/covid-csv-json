```python
import pandas as pd
df_confirmed = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Confirmed.csv")
df_deaths = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Deaths.csv")
```

```python
df_confirmed_wa = df_confirmed.loc[(df_confirmed['Province/State'] == 'Washington')]
df_deaths_wa = df_deaths.loc[(df_deaths['Province/State'] == 'Washington')]
```

```python
df_deaths_wa=df_deaths_wa.rename(index={98: "Deaths"})
df_deaths_wa=df_deaths_wa.transpose()
df_deaths_wa
```

```python
df_confirmed_wa=df_confirmed_wa.rename(index={98: "Confirmed"})
df_confirmed_wa=df_confirmed_wa.transpose()
df_confirmed_wa
```

```python
df_combined_wa=df_confirmed_wa.join(df_deaths_wa)
df_combined_wa
```
