```python
import pandas as pd

def get_covid_dataframe_for_date(date):
    df0 = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports/"+date+".csv");
    df=df0.loc[(df0['Province_State'] == 'Washington')]
    dfs=df[['Admin2','Confirmed','Deaths']]
    dfs=dfs.rename(columns={"Admin2": "County"})
    #dfs.index=dfs['County'] #new
    #dfs.drop(columns='County',inplace=True)
#    dfs['date']=date
    return dfs.set_index('County')

dates=['03-22-2020','03-23-2020','03-24-2020']
frames = [ get_covid_dataframe_for_date(d) for d in dates ]
```
